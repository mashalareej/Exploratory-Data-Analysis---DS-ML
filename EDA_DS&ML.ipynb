```python
#data cleaning
# delete the duplicate movies
data = data.drop_duplicates(subset='MOVIES')
data
```

```python
# removing Gross column since it has a large number of missing values
data.drop(columns=['Gross'], inplace=True)
```

```python
data
```

```python
# deleting duplicate rows

data = data.drop_duplicates()
data
```

```python
data.shape
```

```python
# now for YEAR column, we see that there are different kinds of entries, so we will remove the rows which have unique values

import re
data = data[data['YEAR'].astype(str).str.contains(r'\d{4}', regex=True, na=False)]
data
```

```python
# now there is no null value too in YEAR
data['YEAR'].isnull().sum()
```

```python
# now in GENRE
data['GENRE']
```

```python
# we will remove the \n from all the rows
data['GENRE'] = data['GENRE'].str.lstrip('\n')
data
```

```python
# we will remove the null rows from genre
data = data.dropna(subset=['GENRE'])
data
```

```python
# checking for null rows in RATING
data['RATING'].isnull().sum()
```

```python
# we will now calculate mean of the values in RATING column and fill empty/null values with that
mean_rating = data['RATING'].mean()
mean_rating
```

```python
# now filling the null rows with mean_rating
data['RATING'] = data['RATING'].fillna(mean_rating)
data
```

```python
# confirm that there are no null rows in RATING now
print(data['RATING'].isnull().sum())
```

```python
# clean ONE-LINE
data['ONE-LINE']
```

```python
# remove \n from right and left
data['ONE-LINE'] = data['ONE-LINE'].str.lstrip('\n')
data['ONE-LINE'] = data['ONE-LINE'].str.rstrip('\n')
data
```

```python
# number of null one lines
data['ONE-LINE'].isnull().sum()
```

```python
data.head()
```

```python
# remove null/empty rows from STARS columns cuz everyone wants to know about the stars in a movie
data['STARS'].isnull().sum()
```

```python
# remove \n from the STARS column
data['STARS'] = data['STARS'].str.replace('\n', ' ', regex=True)
data
```

```python
# see for null values in votes and runtime
data['VOTES'].isnull().sum()
```

```python
data['RunTime'].isnull().sum()
```

```python
# now we will fill the null values in runtime to mean value of the column and votes to unknown

data['RunTime'] = data['RunTime'].fillna(data['RunTime'].median())
data['VOTES'] = data['VOTES'].fillna('Unknown')
data
```

```python
# Add 'Director: ' in front if it's missing
data['STARS'] = data['STARS'].apply(lambda x: 'Director: Unknown | ' + x if 'Director:' not in x else x)
data
```

```python
#now split the STARS column after first |
data[['Director', 'Stars']] = data['STARS'].str.split('|', n=1, expand=True)
data
```

```python
# delete the STARS column
data.drop('STARS', axis=1, inplace=True)
```

```python
data
```

```python
#reset the index
data.reset_index(drop=True)
```

```python
# univariate analysis

#numeric columns

sns.set_style("whitegrid")
sns.histplot(data['RATING'], kde=True, color="#D291BC", edgecolor="#B15EFF")
plt.title("Rating Distribution", color="#8A2BE2")
plt.xlabel("Rating", color="#D63384")
plt.ylabel("Count", color="#D63384")
plt.xticks(color="#800080")
plt.yticks(color="#800080")
plt.grid(color="#F3C5FF", linestyle='--', linewidth=0.5)
plt.show()

# the graph shows that 400 - 600 movies have nearly 6.8 rating and a few maybe 10 -15 movies have around 9 rating out of 10
```

```python
sns.set_style("whitegrid")
sns.histplot(data['RunTime'], kde=True, color="#FF69B4", edgecolor="#DA70D6")
plt.title("Runtime Distribution", color="#C71585")
plt.xlabel("Runtime (minutes)", color="#DB4CCA")
plt.ylabel("Count", color="#DB4CCA")
plt.xticks(color="#800080")
plt.yticks(color="#800080")
plt.grid(color="#FAD1E6", linestyle='--', linewidth=0.5)
plt.show()

# the graph shows that around 1200 movies have nearly 100 runtime
```

```python
# the categorical columns have many unique values which cannot be plotted clearly on bar graphs

#it shows 501 different genres and their ratings
data.groupby('GENRE')['RATING'].mean().sort_values(ascending=False)
```

```python
# find out different unique directors
data['Director'].unique()
```

```python
# now we find which director has the most rating
data.groupby('Director')['RATING'].mean().sort_values(ascending=False).head(10)
```

```python
# director with most movies
top_directors = data['Director'].value_counts().head(10)
print(top_directors)
```

```python
# find out avg rating of top directors

top_10_directors = top_directors.index
avg_rating_top_directors = data[data['Director'].isin(top_10_directors)].groupby('Director')['RATING'].mean()

plt.figure(figsize=(10, 6))
avg_rating_top_directors.sort_values().plot(
    kind='barh',
    color="#DB4CCA",
    edgecolor="#800080"
)

plt.title('Average Rating of Top 10 Directors', color="#C71585", fontsize=14)
plt.xlabel('Average Rating', color="#DB4CCA")
plt.ylabel('Director', color="#DB4CCA")
plt.xticks(color="#800080")
plt.yticks(color="#800080")
plt.grid(axis='x', linestyle='--', linewidth=0.5, color="#FAD1E6")
plt.tight_layout()
plt.show()
```

```python
# top 10 genres
data['GENRE'].value_counts().head(10).plot(kind='barh', color='#DA70D6')
```

```python
# heatmap: correlation matrix
sns.set_style("white")
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(numeric_only=True), cmap='RdPu', annot=True, fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=14)
plt.show()
```

```python
# pairplot of rating, votes, runtime

# Create a 'Rating_Level' column for hue
data['Rating_Level'] = pd.cut(data['RATING'], bins=[0, 5, 7, 10], labels=['Low', 'Medium', 'High'])

# Define custom palette
custom_palette = {'Low': '#FF69B4', 'Medium': '#DA70D6', 'High': '#D8BFD8'}

sns.set_style("whitegrid")
sns.pairplot(data[['RATING', 'VOTES', 'RunTime', 'Rating_Level']],
             kind='scatter',
             hue='Rating_Level',
             palette=custom_palette,
             plot_kws={'alpha': 0.7})

plt.suptitle('Pairwise Plot by Rating Levels', fontsize=14, y=1.02)
plt.show()
```
